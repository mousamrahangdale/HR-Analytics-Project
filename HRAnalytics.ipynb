```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE

# Select only important columns
important_cols = [
    'age','gendercode','maritaldesc','racedesc','title','businessunit','employeetype',
    'departmenttype','division','payzone','employeeclassification','jobfunctiondescription',
    'location','performancescore','currentemployeerating','engagementscore',
    'satisfactionscore','worklifebalancescore','trainingtype','trainingoutcome',
    'trainingduration','trainingcost','supervisor','attrition'
]
df = df[important_cols]

# Encode categorical columns
cat_cols = df.select_dtypes(include='object').columns
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# Features and target
X = df.drop('attrition', axis=1)
y = df['attrition']



```

```python
# Features and target
X = df.drop('attrition', axis=1)
y = df['attrition']
```

```python
# Handle imbalanced data
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# Random Forest Model
rf = RandomForestClassifier(random_state=42)
param_grid = {
     'n_estimators':[300, 400, 500],
    'max_depth':[None, 15, 20],
    'min_samples_split':[2, 4, 6],
    'min_samples_leaf':[1, 2, 3],
    'class_weight':['balanced']
}

grid = GridSearchCV(rf, param_grid, cv=5, scoring='f1', n_jobs=-1)
grid.fit(X_train, y_train)
best_model = grid.best_estimator_
```

```python
# Prediction & Evaluation with custom threshold
y_prob = best_model.predict_proba(X_test)[:,1]
y_pred = (y_prob >= 0.4).astype(int) 

#y_pred = best_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Get feature importances
import numpy as np
importances = best_model.feature_importances_
feature_names = X.columns
plt.figure(figsize=(12,8))
norm = importances / importances.max()
colors = plt.cm.viridis(norm)  #  'plasma', 'coolwarm', 'cividis'
sns.barplot(x=importances,y=feature_names, palette=colors)
plt.title('Feature Importance for Employee Attrition', fontsize=16)
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()
```

```python
# Ensure the same order and presence of columns as during training
X_pred = df[X_train.columns]  
df['predicted_attrition'] = (best_model.predict_proba(X_pred)[:,1] >= 0.4).astype(int)
```
